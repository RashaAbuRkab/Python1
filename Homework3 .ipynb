````python
##
# This program is a password strength checker that check how strong is your password. 
#

### Create a tuple that contains all password policy feedback massage/s.
# @param index the index of the feedback massage
# @return the feedback massage as string
#
def passwordPolicy(index):
    ##return a selected feedback massage from the tuple that contains all password policy feedback massage/s.
    passPolicyFeedback=('Must be at least 8 characters long.', 
                        'Must contain a capital character.',
                        'Must contain a lowercase character.',
                        'Must contain a numeric character.',
                        'Must contain any special character(_,%,!,@,. etc)',
                        'Must not contain more than 4 similar characters.',
                        'Must not found in the common breached passwords')
    return passPolicyFeedback[index]

def passwordStrengthMeter(password , breachedpassword): 
    if (len(password) < 8 or password.isupper() or password.islower() or password.isalpha() or password in breachedpassword ):
        print("your password dose not satisfy the following  policy requirements :")
        if len(password) < 8 :
            print("-" + passwordPolicy(0))
        if password.isupper() :
            print("-" + passwordPolicy(2))        
        if password.islower() :
            print("-" + passwordPolicy(1))
        if password.isalpha()  :
            print("-" + passwordPolicy(3))
        if password.isalnum():
            print("-" + passwordPolicy(4))
        if password in breachedpassword :
            print("-" + passwordPolicy(6))
        for i in password :
            if password.count(i) >= 4 :
                print("-" + passwordPolicy(5))
                break 
    else :
        print("your password is a strong ***password*** ")
def main():
    password = input("Enter your password : ")
    r = open("breachedpass.txt",'r')
    x = r.readlines()
    for i in x :
        breachedpassword = []
        n = i.strip()
        breachedpassword.append(n)
    passwordStrengthMeter(password , breachedpassword)
    r.close()
    
## run the program
main()
````
